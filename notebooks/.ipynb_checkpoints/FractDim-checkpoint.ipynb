```python
from scipy.spatial.distance import pdist


def grass_procaccia(N, locs, r):
    # Algorithm
    # ---------
    # Corr(r) = 2 / N (N-1) * Sum_i<j (Theta( r - d_ij ))
    if N > 15000:
        raise Exception('Potential Mem Overload!')
    
    pair_distances = pdist(locs)
    count = np.sum(np.heaviside(r - pair_distances, np.ones(int(N*(N-1)/2))))
    return 2.0 * count / (N*(N - 1))


def check_border_dist(point, island, r):
    dx = point[0] - island.cx
    dy = point[1] - island.cy
    return island.r - np.sqrt(dx * dx + dy * dy) > r


def corr_dim(M, r, N, island, locs):
    # Uses Grassberger-Procaccia algorithm with limited number of cell calculations
    # M = number of cells
    # N = number of points
    
    # get borders
    
    # from points with border dist > r select M points as cell centers
    possible_centers = [i for i, point in enumerate(locs) if check_border_dist(point, island, r)]
    #print(possible_centers)
    #centers = np.random.choice(possible_centers, size=M, replace=False)
    #print(centers)
    # for each center calculate distances and add if in circle with radius r
    count = 0.
    for i in possible_centers:
        center = locs[i]
        count = np.sum(distance.cdist([center], np.delete(locs, i, axis=0)) < r)
     #   print(count)
    # normalize
    return count / (M * (N-1))
    
```

# Island detection using KMeans and EM Clustering
### KMeans test

```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

X, _ = make_blobs(n_samples=400, centers=3, cluster_std=1.5, random_state=1)
kmeans = KMeans(random_state=12, n_clusters=5)

kmeans.fit(X)

# show data
plt.figure(figsize=(10, 7))
plt.scatter(X[:, 0], X[:, 1], s=10)

# scatter centers
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], s=10, c='red')
```

```python
summed_sq_err = []
n_centers = range(1, 13)
for k in n_centers:
    _kmeans = KMeans(n_clusters=k)
    _kmeans.fit(X)
    summed_sq_err.append(_kmeans.inertia_)
    
cluster_df = pd.DataFrame({'n_cluster': n_centers, 'sse': summed_sq_err})
cluster_df.head()

plt.figure(figsize=(10, 5))
plt.plot(cluster_df['n_cluster'], cluster_df['sse'], 'gx-')
```

### EM algorithm using Gaussian Mixture

```python
from sklearn.mixture import GaussianMixture

gm = GaussianMixture(n_components=78, random_state=0,
                    covariance_type='spherical').fit(cells)
labels = gm.predict(cells)

fig = plt.figure(figsize=(8, 8))

for i in range(78):
    bunch = cells[labels == i]
    plt.scatter(bunch[:, 0], bunch[:, 1], s=1)
    plt.scatter(gm.means_[i][0], gm.means_[i][1], marker='x', color='black')
    sigma = np.sqrt(gm.covariances_[i])
    circle = plt.Circle(gm.means_[i], 2*sigma, color='black', fill=False)
    fig.gca().add_patch(circle)

plt.axis('equal')
plt.show()
```

```python
# analyse single cluster distribution
from scipy.stats import norm

heap = cells[labels == 3]
x = heap[:, 0]
y = heap[:, 1]

# plot histogram and normal distribution
# for each dim of the multivariate p-distribution
fig, ax = plt.subplots(1, 2, figsize=(13, 6), sharey=True)
kwargs = dict(bins=30, alpha=0.5)
ax[0].hist(x, **kwargs, label='x', density=True)
ax[1].hist(y, **kwargs, label='y', density=True)

mux = np.mean(x)
muy = np.mean(y)
sigmax = np.std(x)
sigmay = np.std(y)
confidence = 3 # * sigma
xx = np.linspace(mux - confidence*sigmax, mux + confidence*sigmax, 100)
yy = np.linspace(muy - confidence*sigmay, muy + confidence*sigmay, 100)
ax[0].plot(xx, norm.pdf(xx, mux, sigmax), label='Normal distr.')
ax[1].plot(yy, norm.pdf(yy, muy, sigmay), label='Normal distr.')

fig.suptitle('Mulitvariate distribution of cluster islands')
ax[0].set(xlabel='x direction', ylabel='pdf')
ax[1].set(xlabel='y direction')
ax[0].legend()
ax[1].legend()

plt.show()
```

# Box Counting

```python
# Algorithm
import numpy as np


def get_limit(lower, upper, box_size):
    d = upper - lower
    rest = d % box_size
    # needs to be above stop
    # because of np.arange([start, stop)) open interval
    overhang = box_size * 1e-1
    return [lower, upper + box_size - rest + overhang]
    

def count_boxes(points, box_size, limits=None):
    # get limits in each dimension
    if not limits:
        limits = np.asarray([[np.amin(coords_i), np.amax(coords_i)]for coords_i in np.transpose(points)])
    # modify upper limits based on box_side
    limits = [get_limit(lower, upper, box_size) for lower, upper in limits]    
    H, edges = np.histogramdd(points, bins=tuple(np.arange(start, stop, box_size) \
                                            for start, stop in limits))
    count = np.sum(H > 0)
    return count
```

```python
from dataclasses import dataclass

@dataclass
class Island:
    cx: float
    cy: float
    r: float

radii = np.logspace(-1.0, 2.5, 20)
print(radii)
bc_arr = []
for r in radii:
    #n_evals = 50
    bc = np.array([])
    for label, center, var in zip(labels, gm.means_, gm.covariances_):
        #island = Island(center[0], center[1], np.sqrt(var)*2) # 2 sigma area
        island_locs = cells[labels==label]
        bc = np.append(bc, count_boxes(island_locs, r))
        #current_corr_dim = np.append(current_corr_dim, corr_dim(n_evals, r, island_locs.shape[0], island, island_locs))
    bc_arr.append([np.mean(bc[:]), np.std(bc[:])])
    
bc_arr = np.asarray(bc_arr)
```

```python
m, y0 = np.polyfit(np.log10(radii[8:-7]), np.log10(bc_arr[:, 0][8:-7]), deg=1)
print(f'Fit results: {m=:.2f}, {y0=:.2f}')

%matplotlib inline
plt.figure()
#plt.plot(np.log10(r), np.log10(cd), 'x--')
err = 0.5 * (np.log10(bc_arr[:, 0] + bc_arr[:, 1]) - np.log10(bc_arr[:, 0] - bc_arr[:, 1]))
plt.errorbar(np.log10(radii), np.log10(bc_arr[:, 0]), fmt='x--', yerr=err,
            elinewidth=1)
plt.plot(np.log10(radii), m * np.log10(radii) + y0)
plt.xlabel('log r')
plt.ylabel('log BC')
plt.title(' BC Dim. averaged over islands ')
plt.show()
```

```python
from scipy.spatial.distance import cdist

def f(eps, points, y):
    return np.min(cdist([y], points)) < eps
    

def MC(M, points, eps):
    # D = 2
    y = np.random.rand(M, 2) * 60
    
    fvec = np.vectorize(lambda _x0, _x1: f(eps, points, np.asarray([_x0, _x1])))
    count = np.sum(fvec(y[:, 0], y[:, 1])) / M
    
    return count
    
    
scales = np.logspace(0.1, 0.0001, 10)
MCvec = np.vectorize(lambda _e: MC(4000, points, _e))
r = MCvec(scales)

m, y0 = np.polyfit(np.log10(scales), np.log10(r), deg=1)
print(f'Fit results: {m=:.2f}, {y0=:.2f}')

plt.figure()
plt.plot(np.log(scales), np.log(r), 'o-')
plt.axis('equal')
plt.grid(True)
plt.show()
```

```python
# Algorithm
import numpy as np


def get_limit(lower, upper, box_size):
    d = upper - lower
    rest = d % box_size
    # needs to be above stop
    # because of np.arange([start, stop)) open interval
    overhang = box_size * 1e-1
    return [lower, upper + box_size - rest + overhang]
    

def count_boxes(points, box_size, limits=None):
    # get limits in each dimension
    if not limits:
        limits = np.asarray([[np.amin(coords_i), np.amax(coords_i)]for coords_i in np.transpose(points)])
    # modify upper limits based on box_side
    limits = [get_limit(lower, upper, box_size) for lower, upper in limits]    
    H, edges = np.histogramdd(points, bins=tuple(np.arange(start, stop, box_size) \
                                            for start, stop in limits))
    count = np.sum(H > 0)
    return count
```

```python
def BC(locs, state, L):
    cells = locs[state != 0]
    dist = distance.squareform(distance.pdist(cells))
    dist[dist < 1e-7] = 100.
    m = np.min(dist, axis=0)
    xmin = np.mean(m)
    print(xmin)
    
    x0 = np.log10(4*xmin)
    x1 = np.log10(8*xmin)
    scales = np.logspace(x0, x1, 5)

    helper = lambda scale: count_boxes(cells, scale)
    count_boxes_vect = np.vectorize(helper)
    counts = count_boxes_vect(scales)
    
    plt.figure()
    plt.plot(np.log10(1./scales), np.log10(counts), 'o--')
    plt.grid(True)
    plt.show()

    m, y0 = np.polyfit(np.log10(scales[:]), np.log10(counts[:]), deg=1)
    #print(-m)
    #dx = np.log(1./x1) - np.log(1./x0)
    #dy = np.log(count1) - np.log(count0)
    return abs(m)
```
